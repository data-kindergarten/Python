## def 함수
- 함수는 def에 함수 이름을 지정하고 ( )(괄호)와 :(콜론)을 붙인 뒤 다음 줄에 원하는 코드를 작성합니다(함수의 이름을 짓는 방법은 변수와 같습니다).
- 이때 코드는 반드시 들여쓰기를 해야 합니다(들여쓰기 규칙은 if, for, while과 같습니다).

### 함수 생성

```python
# def 함수이름():
#      코드

def hello():
    print("Hello, world")

hello()
```

```python
# 빈 함수 만들기
# 내용이 없는 빈 함수를 만들 때는 코드 부분에 pass를 넣어준다.

def hello():
    pass
```

## 덧셈함수
- 함수에서 값을 받으려면 ()(괄호) 안에 변수 이름(매개변수)을 지정해주면 됨

```python
def 함수이름(매개변수1, 매개변수2):
    코드
```

```python
def add(a, b):
    print(a + b)

add(10, 20)
```

## 함수 결과 반환(return)
- 함수에서 값을 꺼내오는 것(return에 값을 지정하지 않으면 None을 반환)

### 함수 결과 반환(return)

```python
def 함수이름(매개변수):
    return 반환값
```

```python
# return을 사용하면 값을 함수 바깥으로 반환할 수 있고
# 함수에서 나온 값을 변수에 저장할 수 있음

def add(a, b):
    return a + b

x = add(10,20) # add 함수를 호출해서 반환된  a + b 값을 변수 x에 넣어줌
x 
```

### 매개변수는 없지만 반환값이 있는 함수

```python
# 매개변수는 없지만 값만 반환하는 함수를 만들 수 있음

def one():
    return 1

x = one()
x
```

### return 함수 중간에서 빠져나오기

```python
def not_ten(a):
    if a == 10:
        return
    print(a, '입니다.', sep='')

not_ten(5)
```

## 함수에서 값 여러개 반환하기

```python
# return에서 값이나 변수를 ,(콤마)로 구분해서 지정

def 함수이름(매개변수):
    return 반환값1, 반환값2
```

```python
def add_sub(a, b):
    return a + b, a - b

x = add_sub(20, 10)
x
```

## 함수의 호출 과정 이해

```python
def mul(a, b):
    c = a * b
    return c
 
def add(a, b):       # 1. 함수 add로 x, y 호출
    c = a + b        # 2. c = 10 + 20 
    print(c)         # 3. 30 출력
    d = mul(a, b)    # 4. 함수 mul로 x, y 호출한 뒤, d = 10 * 20
    print(d)         # 5. 200 출력
 
x = 10
y = 20
add(x, y)
```
